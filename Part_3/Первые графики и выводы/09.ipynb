```python
import random
import pandas as pd
```

```python

def bus_wait_time():
    return random.randint(0, 10)
```

```python
def train_wait_time():
    return random.randint(0, 5)
```

```python
def total_delay():
    return (
        bus_wait_time()
    + train_wait_time()
    + train_wait_time()
    + bus_wait_time()
    )
```

```python
days = []
for i in range(365 * 5):
    delay = total_delay()
    days.append(delay)
```

```python
pd.DataFrame(days).hist(bins=10)
```
