```python
import pandas as pd
import random
```

```python
def coin_flip():
    score = random.randint(0, 1)
    return score
```

```python
def flips_heads(repeat):
    total = 0
    for i in range(repeat):
        flip = coin_flip()
        total += flip
    return total
```

```python
experiments = []
```

```python
experiments.append(flips_heads(10))
```

```python
df = pd.DataFrame(experiments)
```

```python
df.hist(bins=11, range=(0, 10))
```
