```python
import numpy as np
import matplotlib.pyplot as plt
```

Изобразите стрелкой вектор [75, 15] из кейса об интернет-магазине одежды «НосиВипчик».


```python
vector = np.array([75,15])
```

```python
plt.figure(figsize = (7,7))
plt.axis([0,80,0,80])
plt.plot(0, 0, 'ro')
plt.arrow(0,0,vector[0],vector[1], head_width=1, length_includes_head="True")
plt.xlabel('Quality')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
