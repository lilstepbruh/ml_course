### Шаг 2.2 Проверка данных на аномалии и исправления.

- Отрицательные значения стажа могут быть следствием человеческих ошибок при ручном вводе данных.
- Если данные были перенесены из одной системы в другую, возможно, произошли ошибки при форматировании данных, особенно если системы имели разные правила для хранения или ввода данных.
- Иногда при преобразовании данных (например, из одной единицы измерения в другую) могут возникнуть ошибки.
- Возможны случаи сбоев при сборе данных — автоматические скрипты могли неверно обработать значения, что могло привести к появлению отрицательных значений.

```python
median1 = df[df['days_employed']>0]['days_employed'].median()
median1
```

```python
median2 = df['total_income'].median()
median2
```

```python
def fill_empties1(row):
    if row['days_employed'] < 0 or pd.isna(row['days_employed']):
        return median1
    return row['days_employed']
```

```python
def fill_empties2(row):
    if row['total_income'] < 0 or pd.isna(row['total_income']):
        return median2
    return row['total_income']
```

```python
df[df['days_employed']<0]
```

```python
df['days_employed'] = df.apply(fill_empties1,axis=1)
df['total_income'] = df.apply(fill_empties2,axis=1)

df
```

### Шаг 2.3. Изменение типов данных.

```python
def convert_to_int(row):
    if pd.isna(row['total_income']) == False:
        try:
            return int(row['total_income'])
        except ValueError:
            return row['total_income'] 
    return row['total_income']
```

```python
df['total_income'] = df.apply(convert_to_int,axis=1)
df
```

### Шаг 2.6. Категоризация дохода.

```python
def apply_category(value):
    try:
        if value <= 30000:
            return 'E'
        if 30001 <= value <= 50000:
            return 'D'
        if 50001 <= value <= 200000:
            return 'C'
        if 200000 <= value <= 1000000:
            return 'B'
        if 1000000 <= value:
            return 'A'
    except:
        return value
```

```python
df['total_income_category'] = df['total_income'].apply(apply_category)
```

```python
df
```

### Шаг 2.7. Категоризация целей кредита.

```python
pd.Series(df['purpose'].unique())
```

```python
def purpose_category_func(purpose):
    if 'авто' in purpose.lower():
        return 'операции с автомобилем'
    if 'жиль' in purpose.lower() or 'недвиж' in purpose.lower():
        return 'операции с недвижимостью'
    if 'свадьб' in purpose.lower():
        return 'проведение свадьбы'
    if 'образован' in purpose.lower():
        return 'получение образования'
```

```python
df['purpose_category'] = df['purpose'].apply(purpose_category_func)
```
