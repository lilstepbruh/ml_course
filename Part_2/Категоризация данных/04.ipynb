```python
support_log_grouped = data.groupby('type_id').count()
support_log_grouped
```

```python
def alert_group(messages):
    if messages <= 300:
        return 'средний'
    elif 301 <= messages <= 500:
        return 'высокий'
    else:
        return 'критичный'
```

```python
print(alert_group(10),alert_group(450),alert_group(1000))
```

```python
support_log_grouped['alert_group'] = support_log_grouped['Unnamed: 0'].apply(alert_group)
support_log_grouped
```

```python
support_log_grouped.groupby('alert_group').sum()
```
