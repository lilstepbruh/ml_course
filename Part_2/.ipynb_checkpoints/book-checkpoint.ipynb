### Шаг 2.1 Заполнение пропусков

Обнаружены пропуски значений в столбцах `total_income` и `days_employed`. Подсчитаем доли пропусков для данных колонок

```python
print((df['days_employed'].isnull().sum())/len(df['days_employed']))
print((df['total_income'].isnull().sum())/len(df['total_income']))
```

- Пропуски могут возникнуть из-за человеческих ошибок при сборе или вводе данных. 
- В некоторых случаях пользователи могут не знать точные значения или предпочитать не указывать такие данные, как доход или стаж. 
- Пропуски могут быть следствием сбоев в системах сбора и обработки данных. 
- В случае некорректного формата данных (например, текст вместо числа), такие значения могут быть пропущены в процессе очистки данных или обработки, что также приводит к пустым ячейкам.

### Шаг 2.2 Проверка данных на аномалии и исправления.

- Отрицательные значения стажа могут быть следствием человеческих ошибок при ручном вводе данных.
- Если данные были перенесены из одной системы в другую, возможно, произошли ошибки при форматировании данных, особенно если системы имели разные правила для хранения или ввода данных.
- Иногда при преобразовании данных (например, из одной единицы измерения в другую) могут возникнуть ошибки.
- Возможны случаи сбоев при сборе данных — автоматические скрипты могли неверно обработать значения, что могло привести к появлению отрицательных значений.

```python
median = df[df['days_employed']>0]['days_employed'].median()
median
```

```python
def fill_empties(row):
    if row['days_employed'] < 0 or pd.isna(row['days_employed']):
        return median
    return row['days_employed']
```

```python
df[df['days_employed']<0]
```

```python
df['days_employed'] = df.apply(fill_empties,axis=1)
df
```

### Шаг 2.3. Изменение типов данных.

```python
def convert_to_int(row):
    if pd.isna(row['total_income']) == False:
        try:
            return int(row['total_income'])
        except ValueError:
            return row['total_income'] 
    return row['total_income']
```

```python
df['total_income'] = df.apply(convert_to_int,axis=1)
df
```
